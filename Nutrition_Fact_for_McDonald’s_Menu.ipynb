# How much do beverages, like soda or coffee, contribute to the overall caloric intake?

### Beverages contribute to McDonald's meal overall caloric intake

```python
df['Category'].unique()
```

```python
#Beverages : 'Beverages', 'Coffee & Tea','Smoothies & Shakes'
df_pie = df.copy()
type(df_pie)
```

```python
df_pie['Category'] = df_pie['Category'].replace(['Coffee & Tea', 'Smoothies & Shakes'], 'Beverages')
df_pie['Category'].unique()
```

```python
# Grouping category
group_pie = df_pie.groupby('Category')[['Calories']].mean()
group_pie
```

```python
print(plt.style.available)
```

```python
with plt.style.context('ggplot'):
  fig = plt.figure(figsize=(12,8))
  plt.pie(x=group_pie['Calories'], labels=group_pie.index,autopct='%2.f%%',shadow=True)
  plt.title('Meal Calories McDonald')
  fig.savefig('calories_pie.png')
  plt.show()
  plt.close()
```

### Beverages contribute to daily caloric intake

```python
# Daily caloric intake = 2000 calories
bev_contribute = (group_pie.loc['Beverages']['Calories']/2000)*100
bev_contribute = round(bev_contribute,2)
print(f'Beverages contribute {bev_contribute} % to daily caloric intake.')
```

# Does ordered grilled chicken instead of crispy increase a sandwich's nutritional value?

**Australia New Zealand Food Standards Code (FSC)**

Source : http://www.mydailyintake.net/daily-intake-levels/

Energy	8,700 kilojoules

Protein	50 grams

Fat	70 grams

Saturated Fatty Acids	24 grams

Carbohydrates	310 grams

Sugars	90 grams

Sodium (salt)	2.3 grams

Dietary Fibre	30 grams

```python
df_sandwich_crispy= df.loc[(df['Item'].str.contains('Sandwich')) & (df['Item'].str.contains('Crispy'))]
df_sandwich_crispy
```

```python
df_sandwich_crispy_nutrition = df_sandwich_crispy[['Category','Item','Protein','Total Fat','Saturated Fat','Carbohydrates','Sugars','Sodium','Dietary Fiber']]
df_sandwich_crispy_nutrition['Category']=df_sandwich_crispy_nutrition['Category'].replace(['Chicken & Fish'],'Crispy')
df_sandwich_crispy_nutrition
```

```python
df_sandwich_grilled= df.loc[(df['Item'].str.contains('Sandwich')) & (df['Item'].str.contains('Grilled'))]
df_sandwich_grilled
```

```python
df_sandwich_grilled_nutrition = df_sandwich_grilled[['Category','Item','Protein','Total Fat','Saturated Fat','Carbohydrates','Sugars','Sodium','Dietary Fiber']]
df_sandwich_grilled_nutrition['Category']=df_sandwich_grilled_nutrition['Category'].replace(['Chicken & Fish'],'Grilled')
df_sandwich_grilled_nutrition
```

```python
crispy_grilled = [df_sandwich_crispy_nutrition, df_sandwich_grilled_nutrition]
df_crispy_grilled = pd.concat(crispy_grilled)
df_crispy_grilled.set_index('Category',inplace=True)
df_crispy_grilled
```

```python
def normalisasi(x):
    return (x-min(x))/(max(x)-min(x))
```

```python
df_crispy_grilled[['Protein','Total Fat','Saturated Fat','Carbohydrates','Sugars','Sodium','Dietary Fiber']]=df_crispy_grilled[['Protein','Total Fat','Saturated Fat','Carbohydrates','Sugars','Sodium','Dietary Fiber']].apply(normalisasi)
df_crispy_grilled
```

```python
cg_group = df_crispy_grilled.groupby('Category')[['Protein','Total Fat', 'Saturated Fat', 'Carbohydrates', 'Sugars','Sodium','Dietary Fiber']].mean()
cg_group.reset_index(inplace=True)
cg_group.set_index(['Category'],inplace=True)
cg_group
```

```python
import numpy as np
import matplotlib.pyplot as plt

with plt.style.context('ggplot'):
    fig = plt.figure(figsize=(17,8))
    df_crispy_grilled.iloc[0,1:]
    barWidth = 0.3
    x = np.arange(7)
    plt.bar(x, cg_group.loc['Crispy'], width=barWidth, label=' Sandwich Crispy',color='orange')
    plt.bar(x+barWidth, cg_group.loc['Grilled'], width=barWidth, label=' Sandwich Grilled',color='brown')
    
    plt.legend(loc='best')   
    plt.xticks([a+0.15 for a in x],[b for b in cg_group.columns])
    plt.title('Average Chicken Sandwich Crispy vs Grilled')
    plt.xlabel('Sandwich')
    plt.ylabel('Nutritional Value')
    plt.show()
    fig.savefig('Average Chicken Sandwich Crispy vs Grilled.png')
    plt.close()
```

#What is the least number of items could you order from the menu to meet one day's nutritional requirements?

The reference values used for the Daily Intake Guide are based on those provided in the **Australia New Zealand Food Standards Code (FSC)**.
http://www.mydailyintake.net/daily-intake-levels/

Energy	8,700 kilojoules

Protein	50 grams

Fat	70 grams

Saturated Fatty Acids	24 grams

Carbohydrates	310 grams

Sugars	90 grams

Sodium (salt)	2.3 grams

Dietary Fibre	30 grams

```python
df_nutrition = df.copy()
df_nutrition = df_nutrition[['Item','Protein','Total Fat','Saturated Fat','Carbohydrates','Sugars','Sodium','Dietary Fiber']]
df_nutrition
```

```python
df_nutr = df_nutrition.set_index('Item')
nutr = list(df_nutr.columns)
min_nutr = [50,70,24,310,90,2.3,30]
count = []
c = 0
k = 6
h = 1
for i in range(len(nutr)):
    c = 0
    for j in range(k):
        if df_nutr[nutr[i]].nlargest(j).sum() < min_nutr[i]:
            c = c+1
    if h != i:
        count.append(c)
        h = h+1
for i in range(len(nutr)):
    print(df_nutr[nutr[i]].nlargest(int(count[i])),'\nDengan jumlah total',nutr[i],'=',df_nutr[nutr[i]].nlargest(int(count[i])).sum(),'\n')
```
